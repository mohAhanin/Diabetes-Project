# Naive Bayes Guassian`

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size=0.1, random_state=132)

from sklearn.naive_bayes import GaussianNB
gaussian=GaussianNB()

gaussian.fit(xtrain,ytrain)

print("Naive Bayes score: ",gaussian.score(xtest,ytest))

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = gaussian.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix:\n", cm)
cr = classification_report(ytest, y_pred)
print("Classification Report:\n", cr)
accuracy = accuracy_score(ytest, y_pred)
print("Accuracy Score: ", accuracy)
```

best test size

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Initialize variables
best_accuracy = 0
best_split = 0

# Iterate over different splits
for i in range(1, 100):
    test_size = i / 100.0
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=test_size, random_state=132)

    # Train the model
    gaussian = GaussianNB()
    gaussian.fit(xtrain, ytrain)

    # Predict the Test set results
    y_pred = gaussian.predict(xtest)

    # Calculate accuracy
    accuracy = accuracy_score(ytest, y_pred)
    print(f"Test size: {test_size}, Accuracy Score: {accuracy}")

    # Check if this is the best accuracy so far
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_split = test_size

print(f"\nBest Test Size: {best_split}, Best Accuracy Score: {best_accuracy}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize variables
best_accuracy = 0
best_criterion = ''
best_depth = 0
best_split = 0

# Define the list of criteria
criteria = ['gini', 'entropy']

# Iterate over different criteria, splits, and depths
for criterion in criteria:
    for depth in range(1, 41):
        print('****************')
        for i in range(5, 30):
            test_size = i / 100.0
            xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=test_size, random_state=0)

            # Train the model
            model = DecisionTreeClassifier(criterion=criterion, max_depth=depth, random_state=0)
            model.fit(xtrain, ytrain)

            # Predict the Test set results
            y_model_pred = model.predict(xtest)

            # Calculate accuracy
            accuracy = accuracy_score(ytest, y_model_pred)
            print(f"Criterion: {criterion}, Max Depth: {depth}, Test Size: {test_size}, Accuracy Score: {accuracy}")

            # Check if this is the best accuracy so far
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_criterion = criterion
                best_depth = depth
                best_split = test_size

print(f"\nBest Criterion: {best_criterion}, Best Max Depth: {best_depth}, Best Test Size: {best_split}, Best Accuracy Score: {best_accuracy}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize variables
best_accuracy = 0
best_n_estimators = 0
best_criterion = ''
best_depth = 0
best_split = 0

# Define the list of criteria and n_estimators
criteria = ['gini', 'entropy']
n_estimators_list = [10,20,30,40, 50,70,80, 100, 200, 500]

# Iterate over different criteria, splits, depths, and n_estimators
for criterion in criteria:
    for n_estimators in n_estimators_list:
        for depth in range(1, 41):
            for i in range(5, 30):
                test_size = i / 100.0
                xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=test_size, random_state=0)

                # Train the model
                model = RandomForestClassifier(n_estimators=n_estimators, criterion=criterion, max_depth=depth, random_state=0)
                model.fit(xtrain, ytrain)

                # Predict the Test set results
                y_pred = model.predict(xtest)

                # Calculate accuracy
                accuracy = accuracy_score(ytest, y_pred)
                print(f"Criterion: {criterion}, Number of Estimators: {n_estimators}, Max Depth: {depth}, Test Size: {test_size}, Accuracy Score: {accuracy}")

                # Check if this is the best accuracy so far
                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_n_estimators = n_estimators
                    best_criterion = criterion
                    best_depth = depth
                    best_split = test_size

print(f"\nBest Number of Estimators: {best_n_estimators}, Best Criterion: {best_criterion}, Best Max Depth: {best_depth}, Best Test Size: {best_split}, Best Accuracy Score: {best_accuracy}")
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_i = knn.predict(xtest)
    error_rate.append(np.mean(pred_i != ytest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
