# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(ytest,predknn))
```

```python
print(classification_report(ytest,predknn))
```

```python
error_rate = []

# Will take some time
for i in range(1,40):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_i = knn.predict(xtest)
    error_rate.append(np.mean(pred_i != ytest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# NOW WITH K=36
knn = KNeighborsClassifier(n_neighbors=36)

knn.fit(xtrain,ytrain)
predknn2 = knn.predict(xtest)

print('WITH K=36')
print('\n')
print(confusion_matrix(ytest,predknn2))
print('\n')
print(classification_report(ytest,predknn2))
```

```python
# Normalize Features
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_train[:5, :]
```

```python
for k in ('linear', 'poly', 'rbf', 'sigmoid'):
    model = svm.SVC(kernel=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_train)
    print(k)
    print(accuracy_score(y_train, y_pred))
```

```python
# Using the best model
model = svm.SVC(kernel='rbf')
model.fit(X_train, y_train)
```

```python
patient = np.array([[ 1., 150., 70., 45., 0., 40., 1.5, 25]])

# Normalize the data with the values used in the training set
patient = scaler.transform(patient)
print(patient)
predbest=model.predict(patient)
predbest[0]
```

```python
# Accuracy on Testing Set
X_test = scaler.transform(X_test)
y_pred = model.predict(X_test)
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
y_zero = np.zeros(y_test.shape)
print(accuracy_score(y_test, y_zero))
```

```python
# Compute precision, recall and f1 score
from sklearn.metrics import recall_score, precision_score, f1_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision is", precision)
print("Recall is", recall)
print("F1 score is", f1)
```

GridSearch

```python
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]}
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```
